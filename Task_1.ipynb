# **Importing Libraries**

```python
import pandas as pd
import numpy as np
```

# **Loading Dataset**

```python
df = pd.read_csv('/content/titanic.csv')
df.info()
```

# **Checking null values**

```python
print(df.isnull().sum())
df.describe()
```

**Handling missing values in Age column**

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

**Handling missing values in Cabin column**,
*dropped this column due to large number of missing values*

```python
df.drop('Cabin', axis = 1, inplace = True)
```

**Handling missing values in Embarked Column**

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
print(df.isnull().sum())
```

# **Renamed Pclass Column**

```python
df.rename(columns = {'Pclass' : 'PassangerClass'}, inplace = True)
```

```python
df.head()
```

# **Converted Datatype of Age to Integer**

```python
df['Age'] = df['Age'].astype(int)
```
